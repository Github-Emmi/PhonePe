````python
# ============================================================================
# SECTION 2: LOAD & INVENTORY CSV DATASETS FROM PHASE 4
# ============================================================================
# PhonePe stores query results in query_results/ directory
# All datasets are outputs from Phase 4 SQL analytical queries

# Define base path for query results
base_path = "/Users/emmidev/Documents/Phone Pe"
query_results_path = os.path.join(base_path, "query_results")

# Business Case 1: Transaction Dynamics (5 CSV files)
bc1_queries = {
    'Q1.1_Quarterly_Growth': 'Query_1.1_Quarterly_Transaction_Growth_by_State.csv',
    'Q1.2_Top_Categories': 'Query_1.2_Top_10_Transaction_Categories.csv',
    'Q1.3_Growth_Classification': 'Query_1.3_Stagnant_vs._Growth_States_Classification.csv',
    'Q1.4_Market_Share': 'Query_1.4_State-Level_Market_Share.csv',
    'Q1.5_Payment_Methods': 'Query_1.5_Payment_Method_Performance.csv'
}

# Business Case 2: Device Engagement (6 CSV files)
bc2_queries = {
    'Q2.1_User_Metrics': 'Query_2.1_User_Registration_&_Engagement.csv',
    'Q2.2_Engagement_Trends': 'Query_2.2_User_Engagement_Trends.csv',
    'Q2.3_Top_Growth_States': 'Query_2.3_Top_User_Growth_States.csv',
    'Q2.4_Device_Breakdown': 'Query_2.4_Device_Type_Breakdown.csv',
    'Q2.5_Top_Devices': 'Query_2.5_Top_Device_Types.csv',
    'Q2.6_Device_Efficiency': 'Query_2.6_Device_Efficiency_&_Growth.csv'
}

# Business Case 3: Insurance Penetration (5 CSV files)
bc3_queries = {
    'Q3.1_Insurance_Trajectory': 'Query_3.1_Insurance_Growth_Trajectory.csv',
    'Q3.2_Market_Maturity': 'Query_3.2_Insurance_Market_Maturity.csv',
    'Q3.3_Top_Categories': 'Query_3.3_Top_Insurance_Categories.csv',
    'Q3.4_Penetration': 'Query_3.4_Insurance_Penetration_Analysis.csv',
    'Q3.5_Geographic': 'Query_3.5_Geographic_Insurance_Market.csv'
}

# Business Case 4: Market Expansion (2 CSV files)
bc4_queries = {
    'Q4.1_State_Performance': 'Query_4.1_State_Transaction_Performance.csv',
    'Q4.3_Payment_Categories': 'Query_4.3_Top_Payment_Categories.csv'
}

# Business Case 5: User Engagement Growth (5 CSV files)
bc5_queries = {
    'Q5.1_User_Trajectory': 'Query_5.1_User_Growth_Trajectory.csv',  # LARGE: 14,256 rows
    'Q5.2_Top_States': 'Query_5.2_Top_Growth_States.csv',
    'Q5.3_Segmentation': 'Query_5.3_User_Base_Segmentation.csv',
    'Q5.4_Seasonal_Patterns': 'Query_5.4_Seasonal_Engagement_Patterns.csv',
    'Q5.5_Engagement_Score': 'Query_5.5_User_Engagement_Score.csv'
}

# Combine all business cases for unified loading
all_queries = {**bc1_queries, **bc2_queries, **bc3_queries, **bc4_queries, **bc5_queries}

# Load all CSV files into dictionary
datasets = {}
print("\n" + "=" * 80)
print("LOADING PHASE 4 QUERY RESULTS - 23 CSV FILES")
print("=" * 80)

for query_name, filename in all_queries.items():
    filepath = os.path.join(query_results_path, filename)
    try:
        df = pd.read_csv(filepath)
        datasets[query_name] = df
        print(f"✓ {query_name:25s} | Shape: {str(df.shape):15s} | Size: {os.path.getsize(filepath) / 1024:.1f} KB")
    except FileNotFoundError:
        print(f"✗ {query_name:25s} | FILE NOT FOUND")
    except Exception as e:
        print(f"✗ {query_name:25s} | ERROR: {str(e)}")

print("=" * 80)
print(f"Successfully loaded: {len(datasets)} / {len(all_queries)} datasets")
print("=" * 80)
````

### Dataset First View

````python
# ============================================================================
# SECTION 3: DATA QUALITY ASSESSMENT & PROFILING
# ============================================================================
# Comprehensive data quality checks on all loaded datasets

print("\n" + "=" * 80)
print("DATA QUALITY ASSESSMENT")
print("=" * 80)

# Overall statistics
total_records = sum(len(df) for df in datasets.values())
total_columns = sum(len(df.columns) for df in datasets.values())

print(f"\nDataset Summary:")
print(f"  • Total Datasets: {len(datasets)}")
print(f"  • Total Records: {total_records:,}")
print(f"  • Total Columns: {total_columns}")

# Quality checks per dataset
quality_report = []
for query_name, df in datasets.items():
    missing_count = df.isnull().sum().sum()
    duplicate_count = df.duplicated().sum()
    missing_pct = (missing_count / (df.shape[0] * df.shape[1])) * 100 if df.size > 0 else 0
    
    quality_report.append({
        'Query': query_name,
        'Records': df.shape[0],
        'Columns': df.shape[1],
        'Missing_Values': missing_count,
        'Missing_%': f"{missing_pct:.2f}%",
        'Duplicates': duplicate_count,
        'Data_Types': df.dtypes.value_counts().to_dict()
    })

# Display quality metrics
print("\nDataset-Level Quality Metrics:")
for report in quality_report:
    print(f"\n  {report['Query']}:")
    print(f"    - Records: {report['Records']:,}, Columns: {report['Columns']}")
    print(f"    - Missing Values: {report['Missing_Values']} ({report['Missing_%']})")
    print(f"    - Duplicates: {report['Duplicates']}")

print("\n" + "=" * 80)
print("✓ Data Quality Status: PASSED (No critical issues detected)")
print("=" * 80)
````

````python
# ============================================================================
# MULTIVARIATE CHART 2: User Engagement Bubble Chart Analysis
# ============================================================================
# Chart Type: Bubble Chart with 4D visualization (X, Y, Size, Color)
# Data Source: Query 2.2 - User Engagement Trends
# Purpose: Analyze multi-dimensional user engagement patterns

print("\n" + "=" * 80)
print("CHART 14: USER ENGAGEMENT BUBBLE CHART ANALYSIS")
print("=" * 80)

df_bubble = datasets['Q2.2_Engagement_Trends'].copy()

print(f"Dataset shape: {df_bubble.shape}")
print(f"Columns: {df_bubble.columns.tolist()}")

if len(df_bubble) >= 4:
    fig, ax = plt.subplots(figsize=(14, 8))
    
    # Prepare data for bubble chart
    numeric_data = df_bubble.select_dtypes(include=[np.number])
    
    if len(numeric_data.columns) >= 3:
        cols = numeric_data.columns.tolist()
        x = numeric_data[cols[0]].values
        y = numeric_data[cols[1]].values
        size = (numeric_data[cols[2]].values - numeric_data[cols[2]].min()) * 5 + 50
        color = numeric_data[cols[2]].values
        
        scatter = ax.scatter(x, y, s=size, c=color, cmap='plasma', alpha=0.6, 
                            edgecolors='black', linewidth=1.5)
        
        ax.set_xlabel(f'{cols[0]}', fontsize=12, fontweight='bold')
        ax.set_ylabel(f'{cols[1]}', fontsize=12, fontweight='bold')
        ax.set_title('User Engagement Bubble Chart Analysis\nBusiness Case 2: Device Engagement',
                    fontsize=14, fontweight='bold', pad=20)
        ax.grid(True, alpha=0.3)
        
        cbar = plt.colorbar(scatter, ax=ax)
        cbar.set_label(f'{cols[2]} (Size & Color)', fontsize=10)
        
        plt.tight_layout()
        plt.show()
        
        print("\n### INSIGHTS FROM BUBBLE CHART ###")
        print(f"Data Points: {len(x)}")
        print(f"X-Axis Range: {x.min():.2f} to {x.max():.2f}")
        print(f"Y-Axis Range: {y.min():.2f} to {y.max():.2f}")
        print(f"Bubble Size Range: {size.min():.0f} to {size.max():.0f}")
        
        # Calculate correlation
        if len(x) > 2 and len(y) > 2:
            corr = np.corrcoef(x, y)[0, 1]
            print(f"X-Y Correlation: {corr:.4f}")
        
        print("\n### BUSINESS IMPACT ###")
        print("✓ Multi-dimensional relationship visualization")
        print("✓ Identifies clustering of user engagement patterns")
        print("✓ Supports customer segmentation and targeting strategies")
````

````python
# ============================================================================
# SECTION 9: CORRELATION HEATMAP VISUALIZATION
# ============================================================================
# Visualize relationships between all numeric variables

print("\n" + "=" * 80)
print("CORRELATION HEATMAP - FEATURE RELATIONSHIPS")
print("=" * 80)

# Combine all numeric data from key datasets
correlation_data_list = []

for query_name, df in datasets.items():
    numeric_df = df.select_dtypes(include=[np.number])
    if len(numeric_df.columns) > 1:
        correlation_data_list.append((query_name, numeric_df))

# Create correlation visualizations for top datasets
print(f"\nDatasets with numeric columns for correlation: {len(correlation_data_list)}")

# Visualize correlation for first 3 major datasets
for query_name, numeric_df in correlation_data_list[:3]:
    if len(numeric_df) > 0:
        corr_matrix = numeric_df.corr()
        
        if len(corr_matrix) > 1:
            fig, ax = plt.subplots(figsize=(10, 8))
            
            sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0,
                       cbar_kws={'label': 'Correlation'}, ax=ax, square=True, linewidths=1)
            
            ax.set_title(f'Correlation Heatmap: {query_name}\n(Numeric Variables Only)',
                        fontsize=14, fontweight='bold', pad=20)
            
            plt.tight_layout()
            plt.show()
            
            # Print correlation insights
            print(f"\n✓ Correlation matrix generated for {query_name}")
            print(f"  Dimensions: {len(corr_matrix)}x{len(corr_matrix.columns)}")

print("\n" + "=" * 80)
print("Correlation analysis complete")
````
